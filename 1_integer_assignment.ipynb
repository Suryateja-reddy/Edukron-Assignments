Q6:Check if 97 is a prime number

```python
num = 97
is_prime = True

if num < 2:
    is_prime = False
else:
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            is_prime = False
            break

print(is_prime)
```
